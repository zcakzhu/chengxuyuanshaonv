## 1. General Statistics

```python
num_imps = len(train)
num_clicks = np.sum(train['click'])
CTR = num_clicks/num_imps
CPM = np.sum(train.payprice)/num_imps
eCPC = np.sum(train.payprice)/num_clicks

print("number of impressions: %s\n" % num_imps)
print("number of clicks: %s\n" % num_clicks)
print("CTR: %s\n" % CTR)
print("average cost per impression: %s\n" % CPM)
print("average eCPC: %s\n" % eCPC)
```

## 2. Is the data balanced?

```python
train.click.value_counts()
print('percentage of click:', round(sum(train.click)/len(train.click),5), '%')
print('percentage of unclick:', round(1-sum(train.click)/len(train.click),5), '%')
```

## 3. How does varialbes influence CTR?

```python
def variable(x):
    IMPS = train.groupby(x)['click'].size().astype(int)
    CLICKS = train.groupby(x)['click'].sum().astype(int)
    COST = round(train.groupby(x)['payprice'].sum()/1000,2)
    CTR1 = round((CLICKS/IMPS),5)
    CPM = round(COST/IMPS*1000,2)
    eCPC = round(COST/CLICKS,2)

    stats = pd.concat([IMPS, CLICKS, COST, CTR1, CPM, eCPC], axis=1)
    stats.columns = ['Impressions','Clicks','Cost','CTR','CPM','eCPC']
    stats.loc['Total',:]= stats.sum(axis=0)

    return stats
   
```
